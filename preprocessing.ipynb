# Treating outliers

Variables with outliers:
- crime_rate (might be skewed)
- n_hot_rooms
- rain_fall

```python
# getting P99(99percentile)
np.percentile(df.n_hot_rooms,[99])
# the output is in array form
```

```python
np.percentile(df.n_hot_rooms, [99])[0]
```

```python
uValue = np.percentile(df.n_hot_rooms, [99])[0]
```

```python
# printing rows having value more than P99
df[(df.n_hot_rooms > uValue)]
```

```python
# imputing values > 3*uValue with 3*uValue
df.n_hot_rooms[(df.n_hot_rooms > 3*uValue)] = 3 * uValue
```

```python
# printing rows having value more than P99 after treating outliers
df[(df.n_hot_rooms > uValue)]
```

```python
# getting P1(1percentile)
lValue = np.percentile(df.n_hot_rooms,[1])[0]
```

```python
df[(df.n_hot_rooms  < lValue)]
# there is no lowwer outliers in n_hot_rooms, so no treatment required
```

```python
# lowwer outliers in rainfall
# getting P1(1percentile)
np.percentile(df.rainfall, [1])[0]
```

```python
lValue = np.percentile(df.rainfall, [1])[0]
```

```python
# printing rows having value more than P1
df[(df.rainfall < lValue)]
```

```python
# imputing values < 0.3*lValue with 0.3*lValue
df.rainfall[(df.rainfall < 0.3 * lValue)] = 0.3 * lValue
```

# Treating Null value

```python
df.info()
# we can see that in column n_hos_beds has only 498 count
# whereas others have 506
# so this column has null value
```

```python
#checking for NULL values
print(df.isnull().any())
#null value count
df.isnull().sum()

# from the output we can see that n_hos_beds has null values
```

```python
#imputing missing value with mean value of the n_hs_beds column
df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```

```python
df.info()
df.isnull().any()
# df.isnull().sum()
```

## Variable transformation

```python
# Transoforming crime_rate variable to make it outlier frozenset
sns.jointplot(x='crime_rate', y = 'price', data = df)
# from the graph it can be said that the crime_rate data shows a logrethmic graph
# To transform it as linear we have to use 'log(1 + x)' function, here 1 is added with x as crime_rate value are concetrated near 0 and log function are irregular near 0.
```

```python
# from the graph it can be said that the crime_rate data shows a logrethmic graph
# To transform it as linear we have to use 'log(1 + x)' function, here 1 is added with x as crime_rate value are concetrated near 0 or are 0 and log(0) is not defined(-infinity).
df['crime_rate'] = np.log(1 + df.crime_rate)
```

```python
sns.jointplot(x='crime_rate', y = 'price', data = df)
# now after transformation, grafh looks somewhat linear
```
